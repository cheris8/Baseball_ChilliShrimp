```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('C:/Users/seungjun/Desktop/baseball/data/batter_for_total_babip.csv')
```

```python
data1 = data.copy()
```

```python
data1
```

```python
df = data1.groupby('P_ID').sum()
```

```python
df
```

```python
df['SF'] = df['PA'] - df['AB'] - df['BB'] - df['HP']
```

```python
df['BABIP'] = ( df['HIT'] - df['HR'] ) / (df['AB'] - df['KK'] - df['HR'] - df['SF'])
```

```python
df
```

```python
sum(np.isnan(df['BABIP']))
```

```python
sum(np.isinf(df['BABIP']))
```

```python
df['BABIP'] = df['BABIP'].fillna(0)
sum(np.isnan(df['BABIP']))
```

```python
babip = df[['BABIP']]
babip['P_ID'] = babip.index
```

```python
babip = babip.reset_index(drop = True)
```

```python
babip
```

```python
babip.to_csv('C:/Users/seungjun/Desktop/baseball/data/batter_total_babip.csv')
```
