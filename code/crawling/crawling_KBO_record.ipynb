## Crawling - KBO 기록

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
```

```python
# 크롬드라이버 경로 설정
path = '/Users/kimchaehyeong/Downloads/chromedriver'
```

### 선수기록 - 타자 크롤링

```python
# 타자 url
url = 'https://www.koreabaseball.com/Record/Player/HitterBasic/Situation.aspx'

# 크롬드라이버 시작
driver = webdriver.Chrome(path)
driver.get(url)
time.sleep(10)

# 월별 -> 8월 선택
select = Select(driver.find_element_by_xpath('//*[@id="cphContents_cphContents_cphContents_ddlSituation_ddlSituation"]'))
select.select_by_visible_text('월별')
time.sleep(10)
select = Select(driver.find_element_by_xpath('//*[@id="cphContents_cphContents_cphContents_ddlSituationDetail_ddlSituationDetail"]'))
select.select_by_visible_text('8월')
time.sleep(10)

# 크롤링 시작
table_df_list = []

for i in range(1, 10):
    
    # 해당 페이지 테이블 크롤링
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.select('#cphContents_cphContents_cphContents_udpContent > div.record_result > table')
    table_html = str(table)
    table_df = pd.read_html(table_html)
    table_df = table_df[0]
    table_df_list.append(table_df)
    
    time.sleep(10)
    
    # 다음 페이지로 이동
    if i%5==1:
        pagexpath = '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNo2"]'
    elif i%5==2:
        pagexpath = '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNo3"]'
    elif i%5==3:
        pagexpath = '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNo4"]'
    elif i%5==4:
        pagexpath = '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNo5"]'
    else:
        pagexpath = '//*[@id="cphContents_cphContents_cphContents_ucPager_btnNext"]'
    try:
        driver.find_element_by_xpath(pagexpath).click()
    except:
        break
    
    time.sleep(10)

# 크롬드라이버 종료
driver.quit()
```

```python
# 타자 8월 테이블 -> ok
hitter_august = pd.concat(table_df_list)
hitter_august.reset_index(drop=True, inplace=True)
hitter_august
```

```python
hitter_august.to_csv('../../data/kbo_record_hitter_august.csv')
```
