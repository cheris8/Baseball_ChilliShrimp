```python
pitcher_2020_ratio = pd.read_csv("https://raw.githubusercontent.com/njj06135/Baseball_ChilliShrimp/master/data/pitcher_2020_ratio.csv")
```

```python
pitcher_left = pd.merge(pitcher_left, pitcher_2020_ratio, left_on='P_ID', right_on='P_ID', how='left')
```

```python
pitcher_left
```

```python
pitcher_left['HR'] = (pitcher_left['HR_rate']*pitcher_left['pred_inplay_hit'])/(1-pitcher_left['HR_rate'])
```

```python
pitcher_left['HIT'] = pitcher_left['HR'] + pitcher_left['pred_inplay_hit']
```

```python
pitcher_left['H2'] = pitcher_left['HIT']*pitcher_left['H2_rate']
pitcher_left['H3'] = pitcher_left['HIT']*pitcher_left['H3_rate']
```

```python
pitcher_left = pitcher_left.fillna(0)
```

```python
pitcher_left['BB'] = pitcher_left['PA'] * pitcher_left['BB_rate']
pitcher_left['HP'] = pitcher_left['PA'] * pitcher_left['HP_rate']
pitcher_left['SF'] = pitcher_left['PA'] * pitcher_left['SF_rate']

pitcher_left['AB'] = pitcher_left['PA'] - pitcher_left['BB'] - pitcher_left['HP'] - pitcher_left['SF']
```

```python
pitcher_left['AVG'] = pitcher_left['HIT'] / pitcher_left['AB']
pitcher_left['OBP'] = (pitcher_left['HIT'] + pitcher_left['BB'] + pitcher_left['HP']) / pitcher_left['PA']
pitcher_left['SLG'] = ((pitcher_left['HIT'] - pitcher_left['H2'] - pitcher_left['H3'] - pitcher_left['HR']) + 2 * pitcher_left['H2'] + 3 * pitcher_left['H3'] + 4 * pitcher_left['HR']) / pitcher_left['AB']
pitcher_left['OPS'] = pitcher_left['OBP'] + pitcher_left['SLG']
```
